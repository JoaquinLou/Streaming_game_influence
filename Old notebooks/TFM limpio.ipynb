```python
# function to clean twitch df
def clean_twitch_df(x):
    columns_to_drop = ['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 11'] #Identify non necessary columns
    x.drop(columns_to_drop, axis=1, inplace=True) #Drop non necessary columns
    x.columns = x.columns.str.lower().str.replace(" ","_") #Clean column names, lowercase and replace spaces with "_"
    x['game'] = x['game'].str.lower() #Standarize game names
    print('Clean done!')
```

```python
# Create a function with clean parameters
def year_sales(year):
    url = str('http://www.vgchartz.com/yearly/' + str(year) + '/Global/')
    response = requests.get(url)
    tables = pd.read_html(url)
    sales_df = pd.DataFrame(tables[1])
    columns = ["Pos", "Game.1", "Game.2"]
    sales_df = sales_df.drop(columns, axis=1)
    sales_df.columns = sales_df.columns.str.lower().str.replace(" ","_")
    sales_df['game'] = sales_df['game'].str.lower()
    sales_df['game'] = sales_df['game'].str.replace(r"\(\s*\+?(-?\d+)\s*\)", '');
    sales_df['platform'] = sales_df['game'].apply(lambda st: st[st.find("(")+1:st.find(")")])
```
