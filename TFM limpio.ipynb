```python
# function to clean twitch df
def clean_twitch_df(x):
    columns_to_drop = ['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 11'] #Identify non necessary columns
    x.drop(columns_to_drop, axis=1, inplace=True) #Drop non necessary columns
    x.columns = x.columns.str.lower().str.replace(" ","_") #Clean column names, lowercase and replace spaces with "_"
    x['game'] = x['game'].str.lower() #Standarize game names
    print('Clean done!')
```

# Game sales data

```python
# bring in vg chart data from url
url = 'http://www.vgchartz.com/yearly/2018/Global/'

response = requests.get(url)
```

```python
# make sure the site is responsive 
response.status_code
```

```python
# define the tables variable
tables = pd.read_html(url)
```

```python
# print out table from site using pandas
tables[1]
```

```python
# convert list object to a dataframe 
sales_2018 = pd.DataFrame(tables[1])
sales_2018.head()
```

```python
# look for unnecessary columns
sales_2018.columns
columns = ["Pos", "Game.1", "Game.2"]
```

```python
# take out columns that are not needed for analysis
sales_2018 = sales_2018.drop(columns, axis=1)
sales_2018.head()
```

```python
sales_2018.columns = sales_2018.columns.str.lower().str.replace(" ","_")
sales_2018.head()
```

```python
sales_2018['game'] = sales_2018['game'].str.lower()
sales_2018.head()
```

```python
sales_2018['game'] = sales_2018['game'].str.replace(r"\(\s*\+?(-?\d+)\s*\)", '')
sales_2018.head()
```

```python
sales_2018['platform1'] = sales_2018['game'].apply(lambda st: st[st.find("(")+1:st.find(")")])
sales_2018.head()
```
